Data cleaning manually

```python
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/IPL_Uncleaned_Dataset.xlsx")
df.head()
```

Data cleaning automatically

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/IPL_Uncleaned_Dataset.xlsx")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_excel("/content/drive/MyDrive/IPL_Cleaned_Dataset.xlsx", index=False)
```

subplot

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/drive/MyDrive/IPL_Cleaned_Dataset.xlsx')


fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].hist(df['Runs_Team1'].dropna())
axes[0, 0].set_title('Runs Team1 Distribution')


axes[0, 1].hist(df['Runs_Team2'].dropna())
axes[0, 1].set_title('Runs Team2 Distribution')


df['Team1'].value_counts().plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Team1 Match Count')


df['Winner'].value_counts().plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Winner Count')

plt.tight_layout()
plt.show()
```

seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
df = pd.read_excel('/content/drive/MyDrive/IPL_Uncleaned_Dataset.xlsx')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Venue',
    y='Runs_Team1',
    estimator='mean'
)

plt.title('Average Runs by Venue')
plt.xlabel('Venue')
plt.ylabel('Average Runs')
plt.xticks(rotation=45)

plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.set_palette("Greens")

plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Venue", y="Runs_Team1")

plt.title("Average Runs by Venue")
plt.xticks(rotation=45)
plt.show()
```
